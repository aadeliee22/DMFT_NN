```python
import numpy as np
from scipy.integrate import simps, trapz
import matplotlib.pyplot as plt
```

$$
\begin{align}
G(i\omega_n) = \int_{-\infty}^{\infty} \frac{d \omega'}{2\pi} \frac{A(\omega')}{i\omega_n-\omega'}\\
G(\tau) = \frac{1}{\beta} \sum_n e^{-i\omega_n\tau}G(i\omega_n)
\end{align}
$$

```python
D = 1.
T = 0.001
beta = int(1./T)
N = 1000
if T == 0.001:
    add_dir = '_0.001'
    U_c1, U_c2 = 2.2, 2.59
    U = np.array([1.50, 2.20, 2.40, 2.58, 2.59, 2.60, 2.80, 3.50])
elif T == 0.01:
    add_dir = ''
    U_c1, U_c2 = 2.2, 2.37
    U = np.array([1.50, 2.00, 2.20, 2.36, 2.37, 2.40, 2.70, 3.50])
    
directory = ''
w_len = len(np.loadtxt(f'.{directory}/1to4{add_dir}/Bethe-{2.00:.2f}_solution.dat', \
                       unpack = True, dtype = 'complex128')[0])
x = np.zeros((len(U), w_len), dtype = 'float64')
omega = np.pi/beta * (2*np.arange(N+1)+1) # w_n

for i, u in enumerate(U):
    w, A_w, G_w, S_w = np.loadtxt(f'.{directory}/1to4{add_dir}/Bethe-{u:.2f}_solution.dat', \
                                  unpack = True, dtype = 'complex128')
    x[i] = A_w.real.copy()
for i in range (np.shape(x)[0]):
    plt.title(f'beta = {beta}')
    plt.plot(w.real, x[i], lw = 1)
```

```python
for i in range (np.shape(x)[0]):
    print(abs(1 - simps(x[i], w.real)), abs(1 - trapz(x[i], w.real)))
    # <-- simps    trapz-->
```

```python
def G(x):
    N = len(omega)
    return np.array([simps(x/(1j*omega[i]-w.real), w.real) for i in range (N)])
```

```python
G_omega = np.zeros((len(x), len(omega)), dtype = 'complex128')

for i, u in enumerate(U):
    w, A_w, G_w, S_w = np.loadtxt(f'.{directory}/1to4{add_dir}/Bethe-{u:.2f}_solution.dat', \
                                  unpack = True, dtype = 'complex128')
    x[i] = A_w.real.copy()
    G_omega[i] = G(x[i])
    np.savetxt(f'./Gw/Gw_{beta}_{u:.2f}.dat', np.array([omega.real, G_omega[i].real, G_omega[i].imag]).T)
```
