```python
import numpy as np
from scipy.integrate import simps, trapz
import matplotlib.pyplot as plt
import os.path as path
```

```python
up_num = 69
down_num = 66
U = np.array([0.001*i for i in range(500, 5001)])
docc11 = np.zeros(up_num, dtype='complex128')
docc21 = np.zeros(down_num, dtype='complex128')

directory = '/home/hyejin/Desktop/my_git/Hubbard_NN/1_dmftsolver/NRG/B_14_2.4'
chiNN1 = np.zeros(up_num, dtype='complex128')
chiSS1 = np.zeros(up_num, dtype='complex128')
UU11 = np.zeros(up_num, dtype = 'float64')
n = 0
for i, u in enumerate(U):
    if path.isfile(f'{directory}/Bethe-{u:.3f}.dat')==False: continue
    
    w, A, G_r, G_i, S_r, S_i, Xn_r, Xn_i, Xs_r, Xs_i = np.loadtxt(f'{directory}/Bethe-{u:.3f}.dat', \
                                                                 unpack=True, dtype = 'float64')
    UU11[n] = u
    chiNN1[n] = simps(Xn_r+1j*Xn_i, w)
    chiSS1[n] = simps(Xs_r+1j*Xs_i, w)
    n+=1
    
directory = '/home/hyejin/Desktop/my_git/Hubbard_NN/1_dmftsolver/NRG/B_41_2.4'
chiNN2 = np.zeros(down_num, dtype='complex128')
chiSS2 = np.zeros(down_num, dtype='complex128')
UU21 = np.zeros(down_num, dtype = 'float64')
n = 0
for i, u in enumerate(U):
    if path.isfile(f'{directory}/Bethe-{u:.3f}.dat')==False: continue
    
    w, A, G_r, G_i, S_r, S_i, Xn_r, Xn_i, Xs_r, Xs_i = np.loadtxt(f'{directory}/Bethe-{u:.3f}.dat', \
                                                                 unpack=True, dtype = 'float64')
    UU21[n] = u
    chiNN2[n] = simps(Xn_r+1j*Xn_i, w)
    chiSS2[n] = simps(Xs_r+1j*Xs_i, w)
    n+=1
    
docc11 = chiNN1-chiSS1
docc21 = chiNN2-chiSS2
plt.plot(UU11, docc11.real, '.-')
plt.plot(UU21, docc21.real, '.-')
plt.ylim(0, 0.04)
plt.xlim(2, 3)
plt.grid(True)
plt.savefig('beta10000.png')
```
