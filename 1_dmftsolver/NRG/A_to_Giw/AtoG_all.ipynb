```python
import numpy as np
from scipy.integrate import simps, trapz
import matplotlib.pyplot as plt

import os.path as path
```

$$
\begin{align}
G(i\omega_n) = \int_{-\infty}^{\infty} \frac{d \omega'}{2\pi} \frac{A(\omega')}{i\omega_n-\omega'}\\
G(\tau) = \frac{1}{\beta} \sum_n e^{-i\omega_n\tau}G(i\omega_n)
\end{align}
$$

```python
def G(x):
    return np.array([simps(x/(1j*omega[i]-w), w) for i in range (N)])

directory = '../NRG'
beta = 10000
N = 10000
w_len = len(np.loadtxt(f'{directory}/B_14_beta{beta:d}/Bethe-{2.000:.3f}.dat', \
                       unpack = True, dtype = 'float64')[0])


omega = np.pi/beta * (2*np.arange(N)+1) # w_n
plt.plot(omega)
```
